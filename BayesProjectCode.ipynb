```python
clean_data = pd.read_csv('cleanedbayestweets.csv')
clean_data.head()
```

```python
#code source: https://neptune.ai/blog/sentiment-analysis-python-textblob-vs-vader-vs-flair

from textblob import TextBlob

polarity = []
sentiment = []

tweets = list(clean_data['nobad'])

for tweet in tweets:
    temp = TextBlob(tweet)
    if temp.sentiment[0]>0:
        polarity.append(1)
    else:
        polarity.append(0)
        
    
    sentiment.append(temp.sentiment[1])

```

```python
clean_data['textblob_polarity'] = polarity
clean_data['textblob_sentiment'] = sentiment
clean_data.head()
```

```python
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer

vader_sent = []

for tweet in tweets:
    analyzer = SentimentIntensityAnalyzer()
    sentence = tweet
    vs = analyzer.polarity_scores(sentence)
    if vs['compound']> 0.5:
        vader_sent.append(1)
    else:
        vader_sent.append(0)
```

```python
clean_data['vader_polarity']= vader_sent
clean_data.head()
```

Textblob Sentiment

```python
blob_train = clean_data.sample(frac=0.67,random_state=1999)
blob_test= clean_data.drop(blob_train.index)

blob_train_pos =  blob_train[blob_train.textblob_polarity == 1]
blob_train_neg = blob_train[blob_train.textblob_polarity == 0]

blob_train_pos = blob_train_pos[['text', 'textblob_polarity']]
blob_train_neg = blob_train_neg[['text', 'textblob_polarity']]

print(blob_train_pos)
print(blob_train_neg)
```

```python
sentences_pos = blob_train_pos['text']
word_list_pos= {}
for sent in sentences_pos:
    splitted = sent.split(' ')
    for s in splitted:
        if s in word_list_pos.keys():
            word_list_pos[s] += 1
        else:
            word_list_pos[s] = 1
```

```python
sentences_neg = blob_train_neg['text']
word_list_neg= {}
for sent in sentences_neg:
    splitted = sent.split(' ')
    for s in splitted:
        if s in word_list_neg.keys():
            word_list_neg[s] += 1
        else:
            word_list_neg[s] = 1
```

```python
pos = pd.DataFrame.from_dict(word_list_pos, orient='index')
pos = pos.rename(columns={0:"positive"})

neg = pd.DataFrame.from_dict(word_list_neg, orient='index')
neg = neg.rename(columns={0:"negative"})

neg['positive'] = pos['positive']

train = neg
train = train.fillna(0)
train
```

```python
# Informed priors
Nreviews = np.array([len(word_list_neg),len(word_list_pos)])
TotalReviews = sum(Nreviews)
priors = Nreviews/TotalReviews 

# Probability of Words given review type

WordSums = train.sum(0) # number of words in each review category

v = len(train.value_counts())# number of distinct words in the data
α = 1.5 # smoothing parameter


Pword = np.zeros([len(train.index), len(train.columns)])
for i in range(len(train.index)):
    for j in range(len(train.columns)):
        Pword[i,j] = (train.iloc[i,j] + α)/(WordSums[j] + v*α)
        
Pword[0:4]
```

### Formatting Test for Blob

```python
blob_test_pos =  blob_test[blob_test.textblob_polarity == 1]
blob_test_neg = blob_test[blob_test.textblob_polarity == 0]

blob_test_pos = blob_test_pos[['text', 'textblob_polarity']]
blob_test_neg = blob_test_neg[['text', 'textblob_polarity']]


sentences_pos_test = blob_test_pos['text']
word_list_pos_test= {}
for sent in sentences_pos_test:
    splitted = sent.split(' ')
    for s in splitted:
        if s in word_list_pos_test.keys():
            word_list_pos_test[s] += 1
        else:
            word_list_pos_test[s] = 1
            
```

```python
sentences_neg_test = blob_test_neg['text']
word_list_neg_test= {}
for sent in sentences_neg:
    splitted = sent.split(' ')
    for s in splitted:
        if s in word_list_neg_test.keys():
            word_list_neg_test[s] += 1
        else:
            word_list_neg_test[s] = 1
```

```python
pos_test = pd.DataFrame.from_dict(word_list_pos_test, orient='index')
pos_test = pos_test.rename(columns={0:"positive"})

neg_test = pd.DataFrame.from_dict(word_list_neg_test, orient='index')
neg_test = neg_test.rename(columns={0:"negative"})

neg_test['positive'] = pos_test['positive']

test = neg_test
test = test.fillna(0)
test
```

```python
posts = []
text_df = blob_test['text']
for i in range(len(text_df)):
    row = text_df.iloc[i]
    row = row.split(' ')

    row = pd.DataFrame(row)
    NewReview = row.index

    WordsIndex = list()
    for i in range(len(NewReview)):
        for j in range(len(test.index)):
            if NewReview[i] == test.index[j].strip():
                WordsIndex.append(j)

    likelihoods = np.ones(len(test.columns))
    for j in range(len(test.columns)):
        for i in WordsIndex:
            likelihoods[j] = likelihoods[j]*Pword[i,j]

    numerator = likelihoods * priors
    postprob = np.round(numerator/numerator.sum(),4)
    postprob

    if postprob[0] == min(postprob):
        posts.append('positive')
    else:
        posts.append('negative')
```

```python
print('True Positive Rate Blob')
true = np.asarray(blob_test['textblob_polarity'])
posts=  np.asarray(posts)

print(np.sum(np.logical_and(posts == 'positive', true == 1))/len(posts))

print('True Negative Rate Blob')
true = np.asarray(blob_test['textblob_polarity'])
posts=  np.asarray(posts)

print(np.sum(np.logical_and(posts == 'negative', true == 0))/len(posts))
```
